```python
# We will be using the dataset called as CAR DETAILS FROM CAR DEKHO.csv
# This dataset contains the details of the cars and we will be using this dataset to predict the selling price of the car
# We will be using the Linear Regression model to predict the selling price of the car
```

```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Importing the dataset
dataset = pd.read_csv('CAR DETAILS FROM CAR DEKHO.csv')

# Checking the dataset
dataset.head()
```

```python
# Checking the shape of the dataset
dataset.shape
```

```python
# Checking the info of the dataset
dataset.info()
```

```python
# Let's check the relationship between the year of purchase of the car and the selling price of the car
sns.scatterplot(x = 'year', y = 'selling_price', data = dataset)
```

```python
# Setting up the Limit of the y-axis for a better idea
sns.scatterplot(x = 'year', y = 'selling_price', data = dataset)
plt.ylim(0, 1000000)
```

```python
# Doing the test train split using sklearn
from sklearn.model_selection import train_test_split
X = dataset[['year']]
y = dataset['selling_price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

# Checking the shape of the train and test data
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Linear regression model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predicting the values
y_pred = regressor.predict(X_test)

print(y_pred)
```

```python
# Plotting the regression line
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Year vs Selling Price (Training set)')
plt.xlabel('Year')
plt.ylabel('Selling Price')
plt.show()

```
